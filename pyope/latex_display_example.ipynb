# PyOPE LaTeX 显示示例

本 notebook 演示如何在 Jupyter 中正确显示 LaTeX 格式的 OPE。

## 导入和设置

```python
from pyope.api import OPE
from pyope.operators import BasisOperator, d
from pyope.constants import One
import sympy as sp
```

## 方法 1：直接显示对象（推荐）

在 Jupyter 中，直接输入对象名会自动调用 `_repr_latex_()` 方法。

```python
# 定义 Virasoro OPE
T = BasisOperator('T', bosonic=True)
c = sp.Symbol('c')
OPE[T, T] = OPE.make([c/2 * One, 0, 2*T, d(T)])

# 直接显示（会自动渲染 LaTeX）
result = OPE(T, T)
result  # ← 这行会显示漂亮的 LaTeX 公式
```
